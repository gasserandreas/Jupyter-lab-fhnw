## Hintergrund

Wie Sie im Theorieteil gelernt haben, bildet eine lineare Abbildung $f: \mathbb{R}^n \to \mathbb{R}^m$ Vektoren aus dem reellen Vektorraum $\mathbb{R}^n$ in den reellen Vektorraum $\mathbb{R}^m$ ab und ist durch eine *Abbildungsmatrix* $\mathbf{A}$ darstellbar:

$$
\mathbf{y} = f(\mathbf{x}) = \mathbf{A} \, \mathbf{x},
$$

wobei $\mathbf{x} \in \mathbb{R}^n$, $\mathbf{y} \in \mathbb{R}^m$ und $\mathbf{A} \in \mathbb{R}^{m \times n}$. In diesem Beispiel soll die geometrische Wirkung linearer Abbildungen mit $\mathbf{A} \in \mathbb{R}^{2\times2}$ betrachtet werden, die von einem zweidimensionalen Ursprungsvektor $\mathbf{x}$ auf einen zweidimensionalen Bildvektor $\mathbf{y}$ abbilden.

## Definition eines Polygons

Als Beispiel betrachten wir ein Polygon, das wir durch die Ortsvektoren seiner Eckpunkte definieren:

```python
V = np.array( [[0,5,5,4,3,3,1,0,0],[0,0,6,7,6,3,3,2,0]] )
V
```

Hierbei sind die $n$ Ortsvektoren der Eckpunkte in der Reihenfolge in der sie verbunden werden müssen horizontal in einer Matrix $\mathbf{V} \in \mathbb{R}^{2\times n}$ gestapelt:

$$
V = \begin{pmatrix} \vec{v}_1 & \vec{v}_2 & \dots & \vec{v}_n \end{pmatrix} = \begin{pmatrix} v_{1,x} & v_{2,x} & \dots & v_{n,x} \\ v_{1,y} & v_{2,y} & \dots & v_{n,y} \end{pmatrix}.
$$

Der Punkt $P_1(0,0)$ wird also mit dem Punkt $P_2(5,0)$ verbunden, der Punkt $P_2(5,0)$ mit dem Punkt $P_3(5,6)$, und so weiter. Wir können das Polygon einfach graphisch darstellen, indem wir die $x$-Komponenten der Vektoren (erste Zeile von $\mathbf{V}$) und die $y$-Komponenten der Vektoren (zweite Zeile von $\mathbf{V}$) als zwei getrennte Listen der `plot`-Funktion von Matplotlib übergeben:

```python
plt.plot( V[0,:], V[1,:], label="Originalbild" )
plt.title("Beispielmenge von Eckpunkten")
plt.legend()
```

## Effiziente Anwendung einer linearen Abbildung

Nun möchten wir die Abbildungsmatrix $\mathbf{A} \in \mathbb{R}^{2\times2}$ einer linearen Abbildung auf jeden einzelnen Eckpunkt des Polygons anwenden und das transformierte Polygon berechnen und visualisieren. Dies kann effizient geschehen, indem wir $\mathbf{A} \in \mathbb{R}^{2\times2}$ direkt auf $\mathbf{V}  \in \mathbb{R}^{2\times n}$ anwenden, wie leicht mit der Definition der Matrixmultiplikation nachgeprüft werden kann.

Beispiel:

```python
A = np.array([[1,0],[0,0]])
plt.plot( V[0,:], V[1,:], label="Originalbild" )
plt.plot( A.dot(V)[0,:], A.dot(V)[1,:], label="Abbildung" )
plt.title("Beispielmenge von Eckpunkten")
plt.legend()
```
