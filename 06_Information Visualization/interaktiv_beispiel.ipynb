```python
# libraries importieren
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
sns.set_theme()
```

```python
# source
wetter = pd.read_csv('./data/meteostat_schweiz2022.csv', 
                    parse_dates=True)
wetter['month'] = (
    pd.to_datetime(wetter['date'])
    .dt.strftime('%m-%b')
)
wetter.to_csv('./data/wetter.csv', index=False)
monthly = (wetter
           .set_index('date')
           .groupby('month')
           .mean()
           .reset_index()
          )
```

```python
dfx = (wetter
       .melt(id_vars='date', 
             value_vars=['tavg', 'tmin', 'tmax'], 
             value_name='temp')
      )
dfx['month'] = (
    pd.to_datetime(dfx['date'])
    .dt.strftime('%b')
)
dfx
```

```python
temps = wetter.melt(id_vars='date', value_vars=['tavg', 'tmin', 'tmax'], value_name='temp')
prcps = wetter.melt(id_vars='date', value_vars=['prcp'], value_name='prcps')
dfx = pd.merge(temps, prcps, on='date')
dfx['month'] = pd.to_datetime(dfx['date']).dt.strftime('%b')
dfx
```

```python
plt.figure(figsize=(10,5))
sns.stripplot(dfx, x='month', y='temp', hue='prcps', linewidth=1)
plt.title('Wetter in ZÃ¼rich, 2022')
plt.legend(fontsize=8, title='Percp', title_fontsize=8)
plt.annotate('Data: meteostat.net CC BY-NC 4.0', xy=(0.76, -0.15), xycoords='axes fraction', fontsize=8)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.swarmplot(dfx, x='month', y='temp', hue='prcps', linewidth=1)
plt.title('Wetter in ZÃ¼rich, 2022')
plt.legend(fontsize=8, title='Percp', title_fontsize=8)
plt.annotate('Data: meteostat.net CC BY-NC 4.0', xy=(0.76, -0.15), xycoords='axes fraction', fontsize=8)
plt.show()
```
