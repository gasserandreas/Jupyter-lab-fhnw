```python
#  Kreisdiagramm
# Daten fr das Kreisdiagramm
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]  # Gren der Sektoren in Prozent

# Farben fr die Sektoren
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# Explodiere den ersten Sektor (optional)
explode = (0, 0, 0, 0)  # Explodiere das erste Segment

# Erstelle das Kreisdiagramm
plt.figure(figsize=(8, 8))  # Gre des Diagramms (optional)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Equal aspect ratio sorgt dafr, dass das Diagramm kreisfrmig ist
plt.axis('equal')

# Diagrammtitel hinzufgen
plt.title('Beispiel Kreisdiagramm')

# Diagramm anzeigen
plt.show()
```
