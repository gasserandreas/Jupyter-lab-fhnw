# Wahrscheinlichkeit für gegeben Bereiche

a. die Wahrscheinlichkeit P (−4 ≤ X ≤ 8) wenn X ∼ N (2, 4)

b. die Wahrscheinlichkeit P(2 ≤ X) wenn X ∼ N(1,9)

c. die Wahrscheinlichkeit P(|X| ≤ 1) wenn X ∼ N(−1,16)

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_normal_distribution(mean, std_def, xValues):
  # create points
  x = np.linspace(mean - 4*std_def, mean + 4*std_def, 100)

  # plot normal distribution
  plt.plot(x, norm.pdf(x, mean, std_def))

  # add axis
  for value in xValues:
    plt.axvline(x=value, color='r', linestyle='--')

  # add title and labels
  plt.title('Normal Distribution')
  plt.xlabel('x')
  plt.ylabel('pdf(x)')
  
  # add grid
  plt.grid(True)

  # show plot
  plt.show()
```

```python
# P (−4 ≤ X ≤ 8) wenn X ∼ N (2, 4)
mean = 2
std_def = 2
a = norm.cdf(8, loc=mean, scale=std_def) - norm.cdf(-4, loc=mean, scale=std_def)
print(a)

plot_normal_distribution(mean, std_def, [-4, 8])
```

```python
# P(2 ≤ X) wenn X ∼ N(1,9)
mean = 1
std_def = np.sqrt(9)

b = 1 - norm.cdf(2, loc=mean, scale=std_def)
print(b)

plot_normal_distribution(mean, std_def, [2])
```

```python
# P(|X| ≤ 1) wenn X ∼ N(−1,16)
mean = -1
std_def = np.sqrt(16)

c = e = norm.cdf(1, loc=-1, scale=4) - norm.cdf(-1, loc=-1, scale=4)
print(c)

plot_normal_distribution(mean, std_def, [-1, 1])
```
