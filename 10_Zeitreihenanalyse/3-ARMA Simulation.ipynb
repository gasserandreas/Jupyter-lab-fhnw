```python
from statsmodels.tsa.arima_process import ArmaProcess
import numpy as np
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
```

**Beispiel:** AR(1)-Modell mit Parameter a1=0.9

```python
# Definition des Prozesses
# Parameter des autoregressiven Modells
ar_par = np.array([0.9])
# Keine Moving Average-Parameter
ma_par = np.array([])
```

```python
# Parameter einpacken
ar = np.r_[1,-ar_par]
ma = np.r_[1,ma_par]
# Modell generieren
ARMA_Modell = ArmaProcess(ar,ma)
# Simulieren
ARMA_sim = ARMA_Modell.generate_sample(1000)
# Anzeigen
fig,axs=plt.subplots(1,3)
fig.set_size_inches(20,8)
axs[0].plot(ARMA_sim)
axs[0].set_title('Zeitreihe')
plot_acf(ARMA_sim,ax=axs[1])
plot_pacf(ARMA_sim,ax=axs[2])
plt.show()
```
